```python
from sklearn.metrics import matthews_corrcoef,confusion_matrix
mcc_scores = {}
for column in test_labels.columns[:-1]:  # Exclude the 'num_labels' column
    mcc = matthews_corrcoef(test_labels[column], df[column])
    mcc_scores[column] = mcc
mcc_scores
```

```python
recall_scores = {}
for column in test_labels.columns[:-1]:  # Exclude the 'num_labels' column
    cf = confusion_matrix(test_labels[column], df[column])
    tp = cf[1][1]
    fn = cf[1][0]
    recall = tp / (tp + fn) if (tp + fn) != 0 else 0
    recall_scores[column] = recall
```
